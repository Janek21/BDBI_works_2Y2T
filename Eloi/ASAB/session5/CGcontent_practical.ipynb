**2.1. Let's have a look at the variation in GC content along the gene body**
    *Let's have a look at the variation in GC content along the gene body Choose an appropriate window size and plot the results per-window. (Use pyplot, dictionaries to store base counts, range funtion for non-overlapping windows. Note that matplotlib has to be installed first for example using conda)*


```python
import matplotlib.pyplot as plt
from math import log10
gene = bp
window = 1000

cg_list = [(gene[pos: pos+window].count('C') + gene[pos: pos+window].count('G'))/window for pos in range(0,len(gene)-window,window)]
plt.title('gc content win:1000')
plt.xlabel('Position (Thousands)')
plt.ylabel('Frequency')
plt.plot(cg_list)
plt.grid(True)
```

 Questions:

Yes there are som high windows up to 50 percent, and lows down to 30 percent
 b) What if you modify the window size
 Values will be much more variation if we reduce, and if we increase window frequency will be more and more close to the overall sequence gc_content mean

**2.2. Use small windows of 100bp to check the variation in GC content along the gene body**

```python

gene= bp;
window=100;
cg_list_100 = [(gene[pos: pos+window].count('C') + gene[pos: pos+window].count('G'))/window for pos in range(0,len(gene)-window,window)]

plt.title('gc content win:100')
plt.xlabel('Position (hundreds)')
plt.ylabel('Frequency')
plt.plot(cg_list_100)
plt.grid(True)

max_peak = max(cg_list)
max_pos = 0
for pos in range(len(cg_list)):
    if cg_list[pos] == max_peak:
        max_pos = pos
print(f'We can see the maximum gc content of: {max_peak} in the interval {max_pos*1000}-{(max_pos*1000)+1000}')

```

```python
from math import log10
gene= bp;
window=1000;
cg_list_log= []

for pos in range(0,len(gene)-window,window):
    gene_seq = gene[pos: pos+window]
    gc_count= (gene_seq.count('C') + gene_seq.count('G'))/window
    at_count = 1 - gc_count
    if gc_count != 0 or at_count != 0:
        log_ratio = log10(gc_count/at_count)
        cg_list_log.append(log_ratio)

plt.axhline(y=0, color='red', linestyle='-')
plt.title('gc content win:1000')
plt.xlabel(f'Position (x{window})')
plt.ylabel('Frequency (log_ratio)')
plt.plot(cg_list_log)
plt.grid(True)
```
